```python
"""                            Data Selection in DataFrame                                  """
```

```python
# DataFrame as a dictionary
```

```python
area = pd.Series({
    'California': 423967,
    'Texas': 695662,
    'New York': 141297,
    'Florida': 170312,
    'Illinois': 149995
})

pop = pd.Series({
    'California': 38332521,
    'Texas': 26448193,
    'New York': 19651127,
    'Florida': 19552860,
    'Illinois': 12882135
})

data = pd.DataFrame({'area': area, 'pop': pop})
data
```

```python
data["area"]
```

```python
data["pop"]
```

```python
data.area
```

```python
data.area is data["area"]
```

```python
data.pop is data["pop"]
```

```python
data["density"] = data["pop"] / data["area"]
data
```

```python
# DataFrame as two-dimensional array
```

```python
data.values
```

```python
data.T
```

```python
data.values[0]
```

```python
data["area"]
```

```python
data.iloc[:3, :2]
```

```python
data.loc[:"Illinois", :"pop"]
```

```python
data.loc[data.density > 100, ["pop", "density"]]
```

```python
data.iloc[0, 2] = 90
data
```

```python
# Additional indexing conventions
```

```python
data["Florida":"Illinois"]
```

```python
data[1:3]
```

```python
data[data.density >100]
```

```python
# DataFrame
data = pd.DataFrame({
    "A":[10, 20, 30, 40],
    "B":[5, 15, 25, 35],
    "C":[100, 200, 300, 400]
}, index = ["x", "y", "z", "t"])
data
```

```python
# 5
data.loc[["y", "z"], ["A", "B"]]
```

```python
# 6
data.iloc[[0, 1], [0, 2]]
```

```python
# 7
data.loc[data.B > 20]
```

```python
data["D"] = data["A"] * 2
data
```

```python
data.loc[["x"], ["C"]] = 999
data
```
