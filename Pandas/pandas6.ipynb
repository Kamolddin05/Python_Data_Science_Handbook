```python
"""                Combining Datasets: Concat and Append              """
```

```python
import pandas as pd
import numpy as np
```

```python
def make_df(cols, ind):
    data = {c: [str(c) + str(i) for i in ind]
            for c in cols}
    return pd.DataFrame(data, ind)

make_df("ABCDEFGHIJKLMNPQRSTVWXYZ", range(24))
```

```python
# >>>---------->> Recall: Concatenation of NumPy Arrays <<----------<<< #
```

```python
x = [1, 2, 3]
y = [4, 5, 6]
z = [7, 8, 9]
np.concatenate([x, y, z])
```

```python
x = [[1, 2],
     [3, 4]]
np.concatenate([x, x], axis=1)
```

```python
# >>>-----------> Simple Concatenation with pd.concat <<------------<<< #
```

```python
ser1 = pd.Series(["A", "B", "C"], index = [1, 2, 3])
ser2 = pd.Series(["D", "E", "F"], index = [4, 5, 6])
pd.concat([ser1, ser2])
```

```python
df1 = make_df("AB", [1, 2])
df2 = make_df("AB", [3 ,4])
print(df1); print(df2); print(pd.concat([df1, df2]))
```

```python
df3 = make_df("AB", [0, 1])
df4 = make_df("CD", [0, 1])
print(df3); print(df4); 
print(pd.concat([df3, df4], axis = 1))
```

```python
# Duplicate indices  >> -- >  Takrorlanuvchi indekslar

```

```python
x = make_df("AB", [0, 1])
y = make_df("AB", [2, 3])
```

```python
y.index = x.index
print(x); print(y); print(pd.concat([x, y]))
```

```python
# Catching the repeats as an error >> -- > Takrorlarni xato sifatida ushlash
```

```python
try:
    pd.concat([x, y], verify_integrity= True)
except ValueError as e:
    print("ValueError:", e)
```

```python
# Ignoring the index >> -- > Indeksni e’tiborsiz qoldirish
```

```python
print(x); print(y); print(pd.concat([x, y], ignore_index=True))
```

```python
# Adding MultiIndex keys >> -- > MultiIndex kalitlarini qo‘shish
```

```python
print(x); print(y); print(pd.concat([x, y], keys=['x', 'y']))
```

```python
#Concatenation with joins >> -- > Join bilan birlashtirish
```

```python
df5 = make_df('ABC', [1, 2])
df6 = make_df('BCD', [3, 4])
print(df5); print(df6); print(pd.concat([df5, df6]))
```

```python
print(pd.concat([df5, df6], join='inner'))
```
