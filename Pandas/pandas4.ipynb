```python
# None: Pythonic missing data
```

```python
import numpy as np
import pandas as pd 
```

```python
vals1 = np.array([1, None, 3, 4])
vals1
```

```python
for dtype in ['object', 'int']:
    print("dtype =", dtype)
    %timeit np.arange(1E6, dtype=dtype).sum()
    print()
```

```python
# NaN: Missing numerical data
```

```python
vals2 = np.array([1, np.nan, 3, 4])
vals2.dtype
```

```python
1 + np.nan    # nan
0 * np.nan    # nan
```

```python
vals2.sum(), vals2.min(), vals2.max()
# (nan, nan, nan)
```

```python
np.nansum(vals2), np.nanmin(vals2), np.nanmax(vals2)
# (8.0, 1.0, 4.0)
```

```python
# Detecting null values
```

```python
data = pd.Series([1, np.nan, "Hello", None])
data.isnull()
```

```python
data[data.notnull()]
```

```python
data.dropna()
```

```python
df = pd.DataFrame([[1, np.nan, 2],
                  [2, 3, 5],
                  [np.nan, 4, 6]])
df
```

```python
df.dropna()
```

```python
df.dropna(axis = "columns")
```

```python
df[3] = np.nan
df
```

```python
df.dropna(axis= "columns", how = "all")
```

```python
df.dropna(axis = "rows", thresh= 3)
```

```python
data = pd.Series([1, np.nan, 2, None, 3], index = list("abcde"))
data
```

```python
data.fillna(0)
```

```python
data.fillna(method = "ffill")
```

```python
data.fillna(method = "bfill")
```

```python
import numpy as np
import pandas as pd
```

```python
s = pd.Series([10, None, 20, np.nan, 30, None])
s
```

```python
df = pd.DataFrame([[1, np.nan, 2],
                   [np.nan, 3, 4],
                   [5, 6, np.nan]]
                   )
nan = df.dropna(how= "any", axis = 0)
print (nan)
```

```python
s = pd.Series([1, np.nan, None, 4]).fillna(value = 0)
s
```

```python
s = pd.Series([1, np.nan, 2, None, 3]).fillna(method = "ffill")
s
```

```python
s = pd.Series([np.nan, 1, None, 2, np.nan]).fillna(method = "bfill")
s
```

```python
df = pd.DataFrame([[np.nan, np.nan, np.nan],
                   [1, np.nan, 3],
                   [np.nan, 2, 3]]).dropna(how= "all")

df
```

```python
df = pd.DataFrame([[np.nan, 1, np.nan],
                   [1, 2, 3],
                   [np.nan, np.nan, 3]]).dropna(thresh= 2)
df
```
